```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import toad
import scorecardpy as sc
```

```python
# 读入数据，以银行信用卡数据为例
data = pd.read_csv('rankingcard.csv')
data = data.drop("Unnamed: 0", axis=1)
```

```python
# toad.detect()
# 数据探查，返回每个特性的EDA报告，包括数据类型、样本数、缺失率、唯一值、平均值、标准差等和各百分比分位数
toad.detect(data)
```

### 2.3 计算每列的Gini值和IV值

```python
# toad.quality()
# 数据探查，输出 IV（信息值）、基尼系数、熵和编号。 每个特征的唯一值。 特征按 IV 降序排列。
# 注意：对于大数据集或高维数据，建议设置“iv_only=True”。
toad.quality(data, 'SeriousDlqin2yrs', iv_only=False)
```

```python
# 缺失值填充
data["NumberOfDependents"].fillna(int(data["NumberOfDependents"].mean()), inplace=True)  
data["MonthlyIncome"].fillna(round(data["MonthlyIncome"].mean(), 3), inplace=True)  
```

```python
# 删除年龄为0,两年预期90天的次数超过90次等异常的数据
data = data[data["age"] != 0]
data = data[data.loc[:,"NumberOfTimes90DaysLate"] < 90]
```

```python
# 划分训练集和测试集
tr,ts = train_test_split(data, test_size=0.25, random_state=450)
tr['type'] = 'train'
ts['type'] = 'test'
```

```python
# toad.select()
# 特征初步筛选
# 根据缺失百分比、IV和皮尔逊相关性（与其他特征）进行初步特征选择
'''
target='SeriousDlqin2yrs':目标值
empyt=0.9: the features with missing percentage larger than 90% are filtered;
iv=0.02: the features with IV smaller than 0.02 are eliminated;
corr=0.7: if two or more features have Pearson correlation larger than 0.7, the ones with lower IV are eliminated;
return_drop=False: if set True, the function returns a list of deleted columns;
exclude=None: input the list of features to be excluded from the algorithm, typically ID column and month column.
'''
selected_data, drop_lst = toad.select(tr, target = 'SeriousDlqin2yrs', empty = 0.5, iv = 0.08, corr = 0.6, 
                                     return_drop=True, exclude=['type'])
```

```python
# 删除了缺失值超过50%，iv值低于0.05， 相关性高于0.7的列
selected_test = ts[selected_data.columns]

print(selected_data.shape)
drop_lst 
```

```python
selected_data.info()
```

### 2.1 确定分箱方法

```python
# c = toad.transform.Combiner() 实例化分箱器
# c.fit() 训练
# c.export() 输出分箱结果
# c.update() 更新分箱设置
# c.transform() 将数据集转换成分箱后的结果
# 初步分箱，支持卡方、决策树、Kmeans、等频、等距
'''
c.fit(dataframe, y = ‘target’, method = ‘chi’, min_samples = None, n_bins = None, empty_separate = False)
y: target variable;
method: the method to apply binning. Suport ‘chi’ (Chi-squared), ‘dt’, (decisin tree), ‘kmeans’ (K-means), ‘quantile’ 
    (by the same percentile), and ‘step’ (by the same step);
min_samples: can be a number or a porportion. Minimum number / porportion of samples required in each bucket;
n_bins: mininum number of buckets. If the number is too large, the algorithm will return the maxinum number of buckets it can get;
empty_separate: whether to seperate the missing values in a bucket. If False, missing values will be put along with the bucket of 
    most close bad rate.
'''

# initialise
c = toad.transform.Combiner()
# Train binning with the selected features from previous; use reliable Chi-squared binning, and control that each bucket has at least 5% sample.
c.fit(selected_data, y = 'SeriousDlqin2yrs', method = 'chi', min_samples = 0.05, empty_separate = False, exclude = ['type']) 
```

```python
# 输出初步的分箱结果
c.export()
```

### 2.4.1 事件率图

```python
# toad.plot.bin_plot() 画分箱后的事件率柱状图，曲线代表bad_rate
# 绘制事件率图，计算IV值, 检查bad_rate是否单调
from toad.plot import bin_plot
feature_columns = selected_data.columns.tolist()
feature_columns.remove('SeriousDlqin2yrs')
feature_columns.remove('type')
# feature_columns.remove('type')
for col in feature_columns:
    # 条形是每个分箱类的比例，线是相应的正样本比例（例如不良率）
    # It's recommended to set 'labels = True' for better visualisation.
    bin_plot(c.transform(selected_data[[col, 'SeriousDlqin2yrs']], labels=True), x=col, target='SeriousDlqin2yrs')
```

### 2.2 确定分组方法(调整分箱)

```python
# 调整分箱，IV值较低的特征，改变其分箱设置，直到IV值达到要求为止
# NumberOfTime60-89DaysPastDueNotWorse 这一列特征的iv值较低，调整分箱
 
# Set new bins
rule = {'NumberOfOpenCreditLinesAndLoans':[3,4,7]}

# Pass new bins
c.update(rule)

# Re-check both in-sample and OOT stability.
bin_plot(c.transform(selected_data[['NumberOfOpenCreditLinesAndLoans','SeriousDlqin2yrs']], labels=True), 
         x='NumberOfOpenCreditLinesAndLoans', target='SeriousDlqin2yrs')
```

```python
# 调整分箱，IV值较低的特征，改变其分箱设置，直到IV值达到要求为止
# DebtRatio 这一列特征的iv值较低，调整分箱
 
# Set new bins
rule = {'DebtRatio':[0.011617953,
#   0.21335751,
#   0.292005983,
  0.425616322,
#   0.651485698,
#   1.967140484
                    ]}

# Pass new bins
c.update(rule)

# Re-check both in-sample and OOT stability.
bin_plot(c.transform(selected_data[['DebtRatio','SeriousDlqin2yrs']], labels=True), 
         x='DebtRatio', target='SeriousDlqin2yrs')
```

```python
# toad.plot.badrate_plot() 画bad_rate曲线
# 查看各列的bad_rate
from toad.plot import badrate_plot

selected_total_data = pd.concat([selected_data, selected_test], axis=0)
for col in feature_columns:
# Check the stability of 'var_d2''s bins across time
#badrate_plot(c.transform(train[[col,'target','month']], labels=True), target='target', x='month', by=col)
#badrate_plot(c.transform(OOT[[col,'target','month']], labels=True), target='target', x='month', by=col)

    badrate_plot(c.transform(selected_total_data[[col,'SeriousDlqin2yrs','type']], labels=True), target='SeriousDlqin2yrs', x='type', 
                 by=col, return_counts=True, return_proportion=False)
    print(col)
'''
如果类之间的差距随着时间的推移变得越来越大，那么一个特征是更可取的——这意味着分类后的类有更大的差异。 
没有线交叉意味着 bin 结果是稳定的。
'''
```

### 2.4.2 事件计数+woe图，计算分组时判断bad_rate是否单调

```python
# 将调整后的分组结果应用在训练数据集上，转换为离散值, labels设置为False，之前将labels设置为True是为了方便可视化
transformed_data = c.transform(selected_data, labels=False)
transformed_data
```

```python
# 将调整后的分组结果应用在测试数据集上，转换为离散值, labels设置为False
transformed_testdata = c.transform(selected_test, labels=False)
transformed_testdata
```

```python
# transer = toad.transform.WOETransformer() 实例化WOE转换器
# transer.fit_transform() 训练数据并转换为WOE值
# transer.transform() 训将数据转换为WOE值
transer = toad.transform.WOETransformer()
# transer.fit_transform() & combiner.transform(). Remember to exclude target
train_woe = transer.fit_transform(transformed_data, selected_data['SeriousDlqin2yrs'], exclude=['SeriousDlqin2yrs', 'type'])
test_woe = transer.transform(transformed_testdata)
```

```python
train_woe
```

```python
test_woe
```

```python
# toad.selection.stepwise() 逐步回归法筛选特征
# 使用逐步回归特征选择
'''
- estimator: the regression model to fit, support 'ols', 'lr', 'lasso', 'ridge'

- direction: stepwise direction, support 'forward', 'backward', 'both' (recommended)

- criterion: selection criteria, support 'aic', 'bic', 'ks', 'auc'

- max_iter: maximum number of iterations

- return_drop: whether to return a list of dropped column names

- exclude: list of column to be from alogorithm, such as ID column and time column.
'''
# Apply stepwise regression on the WOE-transformed data
final_data = toad.selection.stepwise(train_woe,target = 'SeriousDlqin2yrs', estimator='ols', direction = 'both', criterion = 'aic', exclude = ['type'])
final_data
```

```python
final_data = final_data.drop(['type'], axis=1)
final_data
```

```python
#  Place the selected features to test 
final_test = test_woe[final_data.columns]
final_test
```

```python
final_tmp_data = final_data.drop('SeriousDlqin2yrs', axis=1)
transformed_tmp_data = transformed_data.drop('SeriousDlqin2yrs', axis=1) 
subtraction_columns = [i for i in transformed_tmp_data if i not in final_tmp_data]
transformed_tmp_data = transformed_tmp_data.drop(subtraction_columns, axis=1)
for i in final_tmp_data.columns.tolist():
    final_tmp_data.rename(columns={i:i+"_woe"}, inplace=True)
```

```python
df3 = pd.concat([transformed_tmp_data, final_tmp_data], axis=1)

for i in transformed_tmp_data.columns.tolist():
    bins_count_plot_data = df3[i].value_counts()
    woe_plot_data = df3[[i, i+'_woe']].drop_duplicates()
    bins_count_plot_dataframe = pd.DataFrame({i:bins_count_plot_data.index, i+'_count':bins_count_plot_data.values})
    df_new = pd.merge(woe_plot_data, bins_count_plot_dataframe, on=i)
    df_new = df_new.sort_values(by=[i])
    
    fig = plt.figure()
#     plt.xticks(rotation=45)    
    ax1 = fig.add_subplot()
    ax1.bar(x=df_new[i], height=df_new[i + '_count'], color='steelblue', label=i + '_count')
    ax1.set_xlabel('bins')
    ax1.set_ylabel('count')
    
    for a,b,k in zip(df_new[i],df_new[i + '_count'],range(len(df_new[i]))): # zip 函数
        plt.text(a,b+0.01,df_new[i + '_count'][k], ha='center') # plt.text 函数
    
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('woe')
    ax2.plot(df_new[i], df_new[i+'_woe'], 'or-', label=i+'_woe')
    for j, (_x, _y) in enumerate(zip(df_new[i], df_new[i+'_woe'])):
        plt.text(_x, _y, round(df_new[i+'_woe'][j],3), color='black', fontsize=10, )  # 将数值显示在图形上
    
    plt.show()
```

```python
# toad.metrics.PSI() 输入两组数据，检查数据的稳定性
# 输出每个特征的 PSI - 用于检查 WOE 转换特征的测试集的稳定性。
xtr = final_data.drop(["SeriousDlqin2yrs"], axis=1)
ytr = final_data['SeriousDlqin2yrs']
xts = final_test.drop(["SeriousDlqin2yrs"], axis=1)
yts = final_test['SeriousDlqin2yrs']
toad.metrics.PSI(xtr, xts)
```

```python
# 用逻辑回归模型训练模型
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(xtr, ytr)

# Obtain predicted probability for training and testing
pred_train = lr.predict_proba(xtr)[:,1]
pred_train
```

```python
pred_test_may =lr.predict_proba(xts)[:,1]
pred_test_may
```

```python
# 计算KS值和AUC值
from toad.metrics import KS, AUC

print('train KS',KS(pred_train, ytr))
print('train AUC',AUC(pred_train, ytr))
print('test KS ', KS(pred_test_may, yts))
```

```python
# PSI 还可用于衡量预测概率的稳定性

print(toad.metrics.PSI(pred_train, pred_test_may))
```

```python
# 输出分箱预测概率的评价信息，包括每个概率分箱的概率范围、样本数、不良率、KS
# Group the predicted scores in bins with same number of samples in each (i.e. "quantile" binning)
toad.metrics.KS_bucket(pred_train, ytr, bucket=10, method = 'quantile')
```

```python
# card = toad.ScoreCard() 实例化评分卡转换类
# card.fit() 训练数据
# card.export() 输出评分卡
# 转换为标准评分卡，基础分为base_score:750，违约概率翻倍时增加的分数为pdo:60，给定初始概率为base_odds：35
card = toad.ScoreCard(
    combiner = c,
    transer = transer,
    #class_weight = 'balanced',
    #C=0.1,
    #base_score = 750,
    #base_odds = 35 ,
    #pdo = 60,
    #rate = 2
)

card.fit(xtr, ytr)
```
